```python
cube  = [x ** 3 for x in range(-10, 11)]
cube2 = [2 * x ** 3 for x in range(-10, 11)]
plt.plot(range(-10, 11), cube, label='$y=x^3$')
plt.plot(range(-10, 11), cube2, label='$y=2*x^3$')
plt.fill_between(range(-10, 11), cube, cube2, color='green', alpha=0.5, label='place between')
plt.legend(loc=2)
```
```python
plt.show()
```

Давайте посмотрим, насколько сильно различаются между собой объекты классов Malignan и Benign по паре признаков 'mean radius' и 'mean texture', а также по паре признаков 'mean perimeter' и 'mean area'. Для этого нужно изобразить объекты этих классов разными цветами в координатах соответсвующих пар признаков (для работы с цевтом подойдет ListedColormap)  и проанализировать полученные изображения. 

На одном полотне отобразите объекты в координатах 'mean radius' и 'mean texture', а на другом - в координатах 'mean perimeter' и 'mean area'.
```python
from matplotlib.colors import ListedColormap
```

```python
colors = ListedColormap(['yellow', 'violet'])
plt.figure(figsize=(12,4))
plt.subplot(121)
plt.scatter(b_frame.loc[:,'mean radius'], b_frame.loc[:,'mean texture'], c = b_frame.target, 
            cmap = colors)
plt.xlabel('mean radius')
plt.ylabel('mean texture')

plt.subplot(122)
plt.scatter(b_frame.loc[:,'mean perimeter'], b_frame.loc[:,'mean area'], c = b_frame.target, 
            cmap = colors)
plt.xlabel('mean perimeter')
plt.ylabel('mean area')

```
